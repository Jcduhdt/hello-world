```python
import tensorflow as tf
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
import os

#
os.getcwd()
image_raw_data_jpg = tf.gfile.FastGFile('./image/4.jpg','rb').read()

with tf.Session() as sess:
    img_data = tf.image.decode_jpeg(image_raw_data_jpg)#
    plt.figure(1)#
    #print(sess.run(img_data))
    plt.imshow(img_data.eval())#
    plt.show()#
    #.eval()
    
    #
    resized = tf.image.resize_images(img_data,[400,800],method=0)
    img_data1 = np.asarray(resized.eval(), dtype='uint8')#
    plt.figure(2)
    plt.imshow(img_data1)
    plt.show
    
    #
    #size
    croped = tf.image.resize_image_with_crop_or_pad(img_data,600,600)
    padded = tf.image.resize_image_with_crop_or_pad(img_data,2000,2000)
    plt.figure(3)
    plt.imshow(croped.eval())
    plt.show()
    plt.figure(4)
    plt.imshow(padded.eval())
    plt.show()
    
    #
    transposed = tf.image.transpose_image(img_data)
    plt.figure()
    plt.imshow(transposed.eval())
    plt.show()
    
    #
    #[-max_delta,max_delta]
    adjusted = tf.image.random_brightness(img_data,max_delta = 0.5)
    plt.figure()
    plt.imshow(adjusted.eval())
    plt.show()
    
    #
    adjusted1 = tf.image.adjust_hue(img_data,0.1)
    plt.figure()
    plt.imshow(adjusted1.eval())
    plt.show()
```
